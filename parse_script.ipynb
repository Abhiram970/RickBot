```python
import re
import pandas as pd
```

```python
pattern = r'([a-zA-Z\s]+):(.+)'
```

```python
data = {
    'name' : [],
    'line' : []
}
```

```python
with open('Rick_trans.txt', 'rt') as file:
  for line in file.readlines():
    match = re.findall(pattern,line)
    if match:
      name,line = match[0]
      data['name'].append(name)
      data['line'].append(line)
```

```python
df = pd.DataFrame(data)
```
